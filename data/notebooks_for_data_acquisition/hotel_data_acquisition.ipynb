### 1. Import Libraries

```
# For data processing
import pandas as pd

# For scraping a booking website
from bs4 import BeautifulSoup
import requests
import re

# For downloading a csv file (if you use Google Colaboratory)
from google.colab import files
```

### 2. Get Hotels Data

* Send GET request to hotels.com 
* Save the HTML response of a requested page
* Pick up information of each hotel on the page
* Find params for the next page
* Loop until there is no next page or no registered price



```
i = 1
continue_flag = True
hotel_values_list = []

url_base = 'https://www.hotels.com'
search = '/search/search.html'
params = '?resolved-location=CITY%3A726784%3AUNKNOWN%3AUNKNOWN&destination-id=726784&q-destination=Tokyo,%20Tokyo%20(prefecture),%20Japan&q-rooms=1&q-room-0-adults=2&q-room-0-children=0'

while continue_flag == True:
  # Send GET request and save the response
  request = requests.get(url_base + search + params)
  bs = BeautifulSoup(request.text, 'html.parser')

  # Find hotel contents tags
  hotel_list = bs.ol.find_all('li', attrs={'class':'hotel'})

  # Find the params for next page
  params = bs.find(attrs={'class':'pagination'}).a['href']
  pagination_param = re.search(r'pn=\d+', params).group()
  print(pagination_param)

  # Stop the loop when there is no next page
  if i > 1 and pagination_param == 'pn=2':
    break

  for hotel in hotel_list:
    # Find the values for each column of the final output
    try:
      hotel_name = hotel.find(attrs={'class':'p-name'}).get_text()
      hotel_address = hotel.find(attrs={'class':'address'}).get_text()
      hotel_area = hotel.find(attrs={'class':'map-link'}).get_text()
    except:
      hotel_name, hotel_address, hotel_area = None, None, None
    
    try:
      hotel_type = re.search(r'(.*)@(.*)', hotel['data-accommodation-type']).group(2)
    except:
      hotel_type = None

    try:
      # Pick up a discounted price if available
      hotel_price_value = hotel.find(attrs={'class':'price'}).get_text()
      hotel_price_list = [int(n) for n in re.findall(r'\d+', hotel_price_value)]
      hotel_price = min(hotel_price_list)
    except AttributeError:
      # Stop the while loop when venue without price appears
      continue_flag = False
      break
    except:
      hotel_price = None
    
    try:
      # Send GET request to the hotel detail page and find the values of latitude and longitude
      hotel_detail_params = hotel.find(attrs={'class':'p-name'}).a['href']
      request_hotel = requests.get(url_base + hotel_detail_params)
      bs_hotel = BeautifulSoup(request_hotel.text, 'html.parser')
      lat_lon_tuple = re.search(r'center=([-]?\d+\.\d+),([-]?\d+\.\d+)&', bs_hotel.find(attrs={'class':'map-widget'})['style']).group(1,2)
      hotel_lat, hotel_lon = lat_lon_tuple[0], lat_lon_tuple[1]
    except:
      hotel_lat, hotel_lon = None, None
    
    hotel_values_list.append([hotel_name, hotel_type, hotel_address, hotel_area, hotel_lat, hotel_lon, hotel_price])
    
  i += 1
```

### 3. Save the Dateset to CSV File

* Create pandas DataFrame from the hotel date(list)
* Save it as a CSV file and download it to your local storage

```
column_list = ['hotel_name', 'hotel_type', 'hotel_address', 'hotel_area', 'hotel_lat', 'hotel_lon', 'hotel_price($)']
hotel_df = pd.DataFrame(columns=column_list, data=hotel_values_list)
hotel_df.head()
```
